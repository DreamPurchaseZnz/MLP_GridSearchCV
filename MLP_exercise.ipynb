# BUILD MODEL STRUCTURE

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.optimizers import Adam
import numpy as np
```

THE FOLLOWING MODEL SET COMPRISE FOUR VARIES

```python
def model_one(dropout=0.5):
    model = Sequential()
    model.add(Dense(512,input_shape=(1867,),name='layer1'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout)) 
    model.add(Dense(1,name='output'))
    model.compile(loss='mean_squared_error',optimizer='adam')
    return model
```

```python
def model_two(dropout=0.5):
    model = Sequential()
    model.add(Dense(512,input_shape=(1867,),name='layer1'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(512,name='layer2'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(1,name='output'))
    model.compile(loss='mean_squared_error',optimizer='adam')
    return model
```

```python
def model_three(dropout=0.5):
    model = Sequential()
    model.add(Dense(512,input_shape=(1867,),name='layer1'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(512,name='layer2'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(512,name='layer3'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(1,name='output'))
    model.compile(loss='mean_squared_error',optimizer='adam')
    return model
```

```python
def model_four(dropout=0.5):
    model = Sequential()
    model.add(Dense(512,input_shape=(1867,),name='layer1'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(512,name='layer2'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(512,name='layer3'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(512,name='layer4'))
    model.add(Activation('relu'))
    model.add(Dropout(dropout))
    model.add(Dense(1,name='output'))
    model.compile(loss='mean_squared_error',optimizer='adam')
    return model
```
